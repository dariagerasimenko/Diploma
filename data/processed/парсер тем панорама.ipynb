```python
classes = []
```

```python
test = df.link
```

```python
URL = test[129]
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(id="panorama-app")
t = results.find("div", class_="meta")
cluster = t.find('span', class_="badge badge-dark")
res = cluster.text.strip()
print(res)
#classes.append(res)
```

```python
for i in test:
    try: 
        URL = i
        page = requests.get(URL)

        soup = BeautifulSoup(page.content, "html.parser")
        results = soup.find(id="panorama-app")
        t = results.find("div", class_="meta")
        cluster = t.find('span', class_="badge badge-dark")
        res = cluster.text.strip()
        classes.append(res)
    except Exception:
        classes.append('Other')
```

```python
result = pd.concat([df, pd.DataFrame(classes, columns = ['class'])], axis = 1)
```

```python
result = result.rename(columns={'0':'class'})
```
