```python
def about_companies(x):
    i1, i2, i3, i4, i5, i6 = 0, 0, 0, 0, 0, 0
    for i in x:
        if '–ª—É–∫–æ–π–ª' in str(i):
            i5+=1
        if '–≤—Ç–±' in str(i):
            i1+=1    
        if '—Å–±–µ—Ä' in str(i):
            i2+=1
        if '–≥–∞–∑–ø—Ä–æ–º' in str(i):
            i3 +=1
        if '—è–Ω–¥–µ–∫—Å' in str(i):
            i4 +=1
        if '—Ä–æ—Å–Ω–µ—Ñ—Ç—å' in str(i):
            i6 +=1
    for i, j in zip(['–ª—É–∫–æ–π–ª', '–≤—Ç–±', '—Å–±–µ—Ä', '–≥–∞–∑–ø—Ä–æ–º', '—è–Ω–¥–µ–∫—Å', '—Ä–æ—Å–Ω–µ—Ñ—Ç—å'], [i5, i1, i2, i3, i4, i6]):
        print(i + ': ' + str(j))
```

```python
x = list(df['message'])
about_companies(x)
```
