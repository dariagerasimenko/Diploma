# Придумываю таргеты



1. y<sub>1</sub> = shift на 10 минут вперед
2. y<sub>2</sub> = shift разницы на 10 минут вперед
3. y<sub>3</sub> = тренд (-1 или +1) за 10 минут
4. y<sub>4</sub> = % прирост за 10 минут
5. y<sub>5</sub> = shift на 60 минут вперед
6. y<sub>6</sub> = shift разницы на 60 минут вперед
7. y<sub>7</sub> = тренд (-1 или +1) за 60 минут
8. y<sub>8</sub> = % прирост за 60 минут
9. y<sub>9</sub> = shift на 1 день вперед
10. y<sub>10</sub> = shift разницы на 1 день вперед
11. y<sub>11</sub> = тренд (-1 или +1) за 1 день
12. y<sub>12</sub> = % прирост за 1 день

```python
# y_1 - shift на 10 минут вперед
result['avg_price'] = ((result['<OPEN>'] + result['<CLOSE>'] + result['<HIGH>'] + result['<LOW>'])/4)
result['target_1'] = result['avg_price'].shift(-1)
result['ticker shift'] = result['<TICKER>'].shift(-1)
result.loc[result['<TICKER>'] == result['ticker shift'], 'y_1'] = result['target_1']
result.loc[result['<TICKER>'] != result['ticker shift'], 'y_1'] = None
```

```python
# y_2 - shift разницы на 10 минут вперед

result['target_2'] = result['y_1'] - result['avg_price']
result.loc[result['<TICKER>'] == result['ticker shift'], 'y_2'] = result['target_2']
result.loc[result['<TICKER>'] != result['ticker shift'], 'y_2'] = None
```

```python
# y_3 - тренд (-1 или +1)

result.loc[result['y_2'] >= 0, 'y_3'] = 1
result.loc[result['y_2'] < 0, 'y_3'] = -1
```

```python
# y_4 - % прироста на 10 минут вперед

result['y_4'] = result['y_1'] / result['avg_price'] - 1
```

```python
# y_5 - shift на 20 минут вперед

result['target_3'] = result['avg_price'].shift(-2)
result['ticker shift 2'] = result['<TICKER>'].shift(-2)
result.loc[result['<TICKER>'] == result['ticker shift 2'], 'y_5'] = result['target_3']
result.loc[result['<TICKER>'] != result['ticker shift 2'], 'y_5'] = None
```

```python
# y_6 - shift разницы на 20 минут вперед

result['target_4'] = result['y_5'] - result['avg_price']
result.loc[result['<TICKER>'] == result['ticker shift 2'], 'y_6'] = result['target_4']
result.loc[result['<TICKER>'] != result['ticker shift 2'], 'y_6'] = None
```

```python
# y_7 - тренд (-1 или +1) за 20 минут

result.loc[result['y_6'] >= 0, 'y_7'] = 1
result.loc[result['y_6'] < 0, 'y_7'] = -1
```

```python
# y_8 - % прироста на 20 минут вперед

result['y_8'] = result['y_5'] / result['avg_price'] - 1
```

```python
# y_9 - shift на 30 минут вперед

result['target_5'] = result['avg_price'].shift(-3)
result['ticker shift 3'] = result['<TICKER>'].shift(-3)
result.loc[result['<TICKER>'] == result['ticker shift 3'], 'y_9'] = result['target_5']
result.loc[result['<TICKER>'] != result['ticker shift 3'], 'y_9'] = None
```

## Лемматизация

```python
!pip install pymorphy2
import pymorphy2
import re
morph = pymorphy2.MorphAnalyzer()
def lemmatize(text):
    """
  Создаём метод, который лемматизирует текст.
  Аргумент:
  text - текст, который нужно лемматизировать
  Возвращает:
  res - лемматизированный текст
    """
    try:
        words = re.split(r'\W+', text) # разбиваем текст на слова
    except:
        return text
    res = str()
    for word in words:
        p = morph.parse(word)[0]
        res += (p.normal_form) + ' '
    return res
```

```python
result['lem_message'] = result['raw_news'].apply(lambda x: lemmatize(x))
```
